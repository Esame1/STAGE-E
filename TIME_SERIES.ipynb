```python
from statsmodels.tsa.stattools import adfuller
adf_result = adfuller(df_daily['SysLoad']) 
print(f'ADF Statistic: {adf_result[0]}') 
print(f'p-value: {adf_result[1]}') 
print(f'No. of lags used: {adf_result[2]}') 
print(f'No. of observations used: {adf_result[3]}') 
print('Critical Values:')
for k, v in adf_result[4].items(): 
      print(f'{k}: {v}')
```

```python
df_daily.reset_index(inplace=True)
```

```python
# Preparing Data for univariate analysis
df_uni = df_daily[['FullDate','ElecPrice']]
```

```python
# Renaming the columns of the new dataframe
new_df = df_uni.rename(columns={"FullDate": "ds", "ElecPrice": "y"})
```

```python
x_train = new_df.iloc[:2757]
x_test = new_df.iloc[2757:]
```

```python
x_test.shape,x_train.shape
```

```python
x_train.head()
```

```python
# building the model
from fbprophet import Prophet
model = Prophet()
```

```python
# fitting the model
model.fit(x_train)
```

```python
# predicting the test data
future = model.make_future_dataframe(periods=1077, freq= 'D' )
p = model.predict(future)
```

```python
p = model.predict(x_test)
p
```

```python
# Plotting the Graph
model.plot(p)
plt.show()
```

```python
#MAPE
from sklearn.metrics import mean_squared_error
y_true, y_pred = np.array(x_test['y']),np.array(p['yhat'])
mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
round(mape,2)
```

```python
#RMSE
np.sqrt(mean_squared_error(x_test['y'],p['yhat']))
```

```python
model.plot_components(p);
```

```python
#Multivariate analysis

#Creatng a new dataframe from df_daily & renaming the columns
df_daily_mv = df_daily.copy()
df_daily_mv.head()
```

```python
df_daily_mul= df_daily_mv.rename(columns={"FullDate": "ds", "ElecPrice": "y","SysLoad": "add1","Tmax": "add2","GasPrice": "add3"})
df_daily_mul.head()
```

```python
x_train_mv = df_daily_mul.iloc[:2757,:]
x_test_mv = df_daily_mul.iloc[2757:,:]
```

```python
model_mv = Prophet()
model_mv.add_regressor('add1')
model_mv.add_regressor('add2')
model_mv.add_regressor('add3')
model_mv.fit(x_train_mv)
```

```python
pred_mv = model.predict(x_test_mv)
```

```python
future_mv = model_mv.make_future_dataframe(periods=1077,freq='D')
future_mv['add1'] = df_daily_mul['add1'].values
future_mv['add2'] = df_daily_mul['add2'].values
future_mv['add3'] = df_daily_mul['add3'].values
pred_mv= model_mv.predict(future_mv)
```

```python
model_mv.plot(pred_mv)
plt.show()
```

```python
y_true, y_pred = np.array(x_test_mv['y'][-1077:]),np.array(pred_mv['yhat'][-1077:])
mape_mv = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
round(mape_mv,2)
```
